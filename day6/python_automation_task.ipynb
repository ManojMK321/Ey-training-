## 1. argparse - Command Line Automation

```python
import argparse

parser = argparse.ArgumentParser(description='User Greeting Automation')
parser.add_argument('--name', type=str, help='User name')

# Simulating CLI input in notebook
args = parser.parse_args(args=['--name', 'Manoj'])

print(f'Hello {args.name}')
```

## 2. schedule - Task Scheduler

```python
import schedule
import time

def job():
    print('Running automated task...')

schedule.every(2).seconds.do(job)

for i in range(5):
    schedule.run_pending()
    time.sleep(2)
```

## 3. subprocess - System Automation

```python
import subprocess

result = subprocess.run(
    'echo Hello Automation',
    capture_output=True,
    text=True,
    shell=True
)

print(result.stdout)
```

## 4. cron - Linux Scheduler

```python
# Cron is configured in the system terminal
# Step 1: Open terminal
# crontab -e

# Step 2: Add job
# * * * * * /usr/bin/python3 /path/to/script.py

print('Cron jobs run outside Python using system scheduler')
```

## 5. Combined Automation Example

```python
import schedule
import subprocess
import time

def automation_job():
    print('Running job...')
    result = subprocess.run('echo Task executed', capture_output=True, text=True, shell=True)
    print(result.stdout)

schedule.every(3).seconds.do(automation_job)

for i in range(5):
    schedule.run_pending()
    time.sleep(3)
```
