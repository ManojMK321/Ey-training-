Task 1: Identify the Flaws
🔹 1. Why does this table violate 1NF?
Violation of 1NF (First Normal Form):
1NF requires that all columns contain atomic (indivisible) values and no repeating groups.
The culprit column is Prescriptions
It contains multiple values in a single field (e.g., "Amoxicillin, Ibuprofen")
👉 This is a multi-valued attribute, which breaks 1NF.

🔹 2. Why do Student_Name and Doctor_Clinic violate 2NF and 3NF?
Assuming Visit_ID is the primary key:
✅ 2NF Violation:
2NF requires:
Table must be in 1NF ✅
All non-key attributes must depend on the whole primary key
Here:
Student_Name depends on Student_ID, not directly on Visit_ID
Doctor_Clinic depends on Doctor_ID, not directly on Visit_ID
👉 These are partial dependencies, so 2NF is violated
✅ 3NF Violation:
3NF requires:
No transitive dependencies
Here:
Visit_ID → Doctor_ID → Doctor_Clinic
👉 Doctor_Clinic depends on Doctor_ID (not directly on Visit_ID)
So this is a transitive dependency, violating 3NF.

🔹 3. Which normal form is violated if Doctor_ID → Doctor_Clinic?
Given:
Doctor_ID → Doctor_Clinic
Doctor_ID is not a candidate key
👉 This violates 3NF
Because:
A non-key attribute (Doctor_Clinic) depends on another non-key attribute (Doctor_ID)

```python

# RUN THIS CELL TO SETUP THE CHALLENGE DATA
import pandas as pd
import sqlite3
 
conn_challenge = sqlite3.connect(':memory:')
 
challenge_data = {
    "Visit_ID": [5001, 5001, 5002, 5003],
    "Student_ID": [101, 101, 102, 104],
    "Student_Name": ["Alice", "Alice", "Bob", "David"],
    "Doctor_ID": ["DOC_XYZ", "DOC_XYZ", "DOC_ABC", "DOC_XYZ"],
    "Doctor_Name": ["Dr. Evans", "Dr. Evans", "Dr. Green", "Dr. Evans"],
    "Doctor_Clinic": ["General Medicine", "General Medicine", "Sports Med", "General Medicine"],
    "Prescriptions": ["Amoxicillin, Ibuprofen", "Amoxicillin, Ibuprofen", "Bandages", "Vitamin D"]
}
 
df_challenge = pd.DataFrame(challenge_data)
df_challenge.to_sql('Patient_Visits_0NF', conn_challenge, index=False, if_exists='replace')
 
print("--- Challenge Dataset (0NF) ---")
df_challenge
```

```python
# Task 2: Normalize to 1NF with Code
# We need to split the multi-valued Prescriptions column into separate rows.
# ✅ Solution using Pandas:
# # Split the prescriptions column into lists
df_1nf = df_challenge.copy()
df_1nf['Prescriptions'] = df_1nf['Prescriptions'].str.split(', ')

# Explode the list into separate rows
df_1nf = df_1nf.explode('Prescriptions')

# Reset index for clean output
df_1nf = df_1nf.reset_index(drop=True)

print("--- Normalized Dataset (1NF) ---")
df_1nf


# ✅ Result:
# Each row now contains only one prescription
# The table now satisfies 1NF
```
